```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load image from local folder (no Colab upload)

img_path = r"D:\FDIP research paper\cat.jpeg"  # update if your image is in a different folder
if not os.path.exists(img_path):
    raise FileNotFoundError(f"Image not found at: {img_path}")
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
if img is None:
    raise RuntimeError(f"cv2 failed to load the image at: {img_path}")
# Contrast stretching: scale pixel values to the full 0-255 range
min_val = np.min(img)
max_val = np.max(img)
if max_val == min_val:
    contrast_stretched = img.copy()
else:
    contrast_stretched = ((img - min_val) / (max_val - min_val) * 255).astype(np.uint8)
```

```python
# Histogram equalization
hist_equalized = cv2.equalizeHist(img)

# Display images and histograms
plt.figure(figsize=(12, 8))

# Original Image
plt.subplot(3, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis("off") 
```

```python
# Histogram of Original
plt.subplot(3, 2, 2)
plt.hist(img.ravel(), bins=256, range=(0, 256), color='black')
plt.title("Histogram - Original")
```

```python
# Contrast Stretched Image
plt.subplot(3, 2, 3)
plt.imshow(contrast_stretched, cmap='gray')
plt.title("Contrast Stretched")
plt.axis("off")
```

```python
# Histogram of Contrast Stretched
plt.subplot(3, 2, 4)
plt.hist(contrast_stretched.ravel(), bins=256, range=(0, 256), color='black')
plt.title("Histogram - Contrast Stretched")
```

```python
# Histogram Equalized Image
plt.subplot(3, 2, 5)
plt.imshow(hist_equalized, cmap='gray')
plt.title("Histogram Equalized")
plt.axis("off")
```

```python

# Histogram of Equalized Image
plt.subplot(3, 2, 6)
plt.hist(hist_equalized.ravel(), bins=256, range=(0, 256), color='black')
plt.title("Histogram - Equalized")
```
