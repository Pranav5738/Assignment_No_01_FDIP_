```python
try:
    import cv2
except Exception as e:
    raise ImportError("OpenCV (cv2) is required. Install it with: pip install opencv-python") from e
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# Try common filenames and give a clear error if none are found
candidates = [Path(r"D:\FDIP research paper\car.jpg"), Path(r"D:\FDIP research paper\car.jpeg")]
image_path = next((p for p in candidates if p.exists()), None)
if image_path is None:
    raise FileNotFoundError(f"Image file not found. Checked: {[str(p) for p in candidates]}")

img = cv2.imread(str(image_path), cv2.IMREAD_GRAYSCALE)
if img is None:
    raise FileNotFoundError(f"Failed to read image at {image_path!s}. Is it a valid image file?")

plt.figure(figsize=(6,6))
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.title("Original Image")
plt.axis("off")
plt.show()

plt.figure()
plt.hist(img.ravel(), bins=256, range=(0, 256), color='black')
plt.title("Histogram - Original")
plt.show()
```

```python
mean = 0
sigma = 20
gaussian = np.random.normal(mean, sigma, img.shape).astype(np.float32)
gaussian_noisy = img.astype(np.float32) + gaussian
gaussian_noisy = np.clip(gaussian_noisy, 0, 255).astype(np.uint8)

plt.imshow(gaussian_noisy, cmap='gray')
plt.title("Gaussian Noise")
plt.axis("off")
plt.show()

plt.hist(gaussian_noisy.ravel(), bins=256, range=(0, 256), color='black')
plt.title("Histogram - Gaussian Noise")
plt.show()
```

```python
# Salt & Pepper Noise
prob = 0.02
salt_pepper_noisy = img.copy()
rnd = np.random.rand(*img.shape)
salt_pepper_noisy[rnd < prob/2] = 0
salt_pepper_noisy[rnd > 1 - prob/2] = 255

plt.imshow(salt_pepper_noisy, cmap='gray')
plt.title("Salt & Pepper Noise")
plt.axis("off")
plt.show()

plt.hist(salt_pepper_noisy.ravel(), bins=256, range=(0, 256), color='black')
plt.title("Histogram - Salt & Pepper Noise")
plt.show()
```

```python
# Uniform Noise
low, high = -50, 50
uniform = np.random.uniform(low, high, img.shape).astype(np.float32)
uniform_noisy = img.astype(np.float32) + uniform
uniform_noisy = np.clip(uniform_noisy, 0, 255).astype(np.uint8)

plt.imshow(uniform_noisy, cmap='gray')
plt.title("Uniform Noise")
plt.axis("off")
plt.show()

plt.hist(uniform_noisy.ravel(), bins=256, range=(0, 256), color='black')
plt.title("Histogram - Uniform Noise")
plt.show()
```

```python
# Lognormal Noise
lognormal = np.random.lognormal(mean=0.0, sigma=0.4, size=img.shape).astype(np.float32)
lognormal_noisy = img.astype(np.float32) + lognormal
lognormal_noisy = np.clip(lognormal_noisy, 0, 255).astype(np.uint8)

plt.imshow(lognormal_noisy, cmap='gray')
plt.title("Lognormal Noise")
plt.axis("off")
plt.show()

plt.hist(lognormal_noisy.ravel(), bins=256, range=(0, 256), color='black')
plt.title("Histogram - Lognormal Noise")
plt.show()
```

```python
# Rayleigh Noise
rayleigh = np.random.rayleigh(scale=30, size=img.shape).astype(np.float32)
rayleigh_noisy = img.astype(np.float32) + rayleigh
rayleigh_noisy = np.clip(rayleigh_noisy, 0, 255).astype(np.uint8)

plt.imshow(rayleigh_noisy, cmap='gray')
plt.title("Rayleigh Noise")
plt.axis("off")
plt.show()

plt.hist(rayleigh_noisy.ravel(), bins=256, range=(0, 256), color='black')
plt.title("Histogram - Rayleigh Noise")
plt.show()
```

```python
# Exponential Noise
exponential = np.random.exponential(scale=30, size=img.shape).astype(np.float32)
exponential_noisy = img.astype(np.float32) + exponential
exponential_noisy = np.clip(exponential_noisy, 0, 255).astype(np.uint8)

plt.imshow(exponential_noisy, cmap='gray')
plt.title("Exponential Noise")
plt.axis("off")
plt.show()

plt.hist(exponential_noisy.ravel(), bins=256, range=(0, 256), color='black')
plt.title("Histogram - Exponential Noise")
plt.show()
```

```python
# Erlang Noise
erlang = np.random.gamma(shape=2, scale=30, size=img.shape).astype(np.float32)
erlang_noisy = img.astype(np.float32) + erlang
erlang_noisy = np.clip(erlang_noisy, 0, 255).astype(np.uint8)

plt.imshow(erlang_noisy, cmap='gray')
plt.title("Erlang Noise")
plt.axis("off")
plt.show()

plt.hist(erlang_noisy.ravel(), bins=256, range=(0, 256), color='black')
plt.title("Histogram - Erlang Noise")
plt.show()
```
